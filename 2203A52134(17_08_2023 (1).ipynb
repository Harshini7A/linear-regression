```python
import pandas as pd
a=pd.read_csv('/content/Salary_Data.csv')
print(a)
```

```python
x=a.drop(columns=["Salary"],axis=1)
y=a['Salary']
print(x)
print(y)
```

```python
features=a.columns[:-1]
target=a.columns[-1]
print(features)
print(target)
```

```python
import matplotlib.pyplot as plt
for feature in features:
  plt.plot(a[feature],a[target],label=feature)
plt.xlabel('features')
plt.ylabel('target variable')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([2,4,6,8,10])
y=np.array([40000,60000,80000,100000,120000])
m,b=np.polyfit(x,y,1)
y_pred=m*x+b
plt.scatter(x,y,label='Data Points')
plt.plot(x,y_pred,color='red',label='Best Fit Line')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.show()
```
